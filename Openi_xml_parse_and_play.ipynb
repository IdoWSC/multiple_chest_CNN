# OPEN-i Data Reports Analysis

```python
import xml.etree.ElementTree as ET
import os


xml_path = '/Users/Ido/Desktop/project_data/OPEN-i/data_reports'
```

```python
def sorted_by_ind(file_list):
    return sorted(file_list, key=lambda x: int(os.path.splitext(x)[0]))
```

```python
xml_files = [file for file in os.listdir(xml_path) if file.endswith('.xml')]
files_list = [os.path.join(xml_path, file) for file in sorted_by_ind(xml_files)]

for file in files_list[:20]:
    print(file)
```

```python
xmls = []
for file in files_list:
    xmls.append({'path': file, 'xml': ET.parse(file)})
print(xmls[0])
```

```python
print(xmls[0])
```

```python
first_xml = xmls[0]
print(first_xml['xml'].getroot().tag)
```

```python
for xml in (xmls[:1]):
    root = xml['xml'].getroot()
    print('root: ', root.tag)
    print('et to string: ', ET.tostring(root, encoding='utf8').decode('utf8'))
    print('\n')
    
    for child in root:
        print('child: ')
        print ( ' ' * 4, child.tag, child.attrib)
```

```python
for xml in (xmls[:10]):
    root = xml['xml'].getroot()
    for pmc_id in root.iter('pmcId'):
        print('path: {} id : {}'.format(xml['path'], pmc_id.attrib['id']))
    
    for abstract_text in root.iter('AbstractText'):
        print ( ' ' * 4, abstract_text.tag, abstract_text.attrib)
        try:
            print(' ' * 8, abstract_text.text + '\n')
        except TypeError:
            print(' ' * 8, 'None.' + '\n')
    
    for parent_image in root.iter('parentImage'):
        print('parentImage: ', parent_image.attrib['id'])
    print('-' * 20)
    
```

## Search for data with effusion mentioned

```python
effusion_mention = []
effusion_non_mention = []
for xml in xmls:
    root = xml['xml'].getroot()
    found_effusion = False
    for abstract_text in root.iter('AbstractText'):
        try:
            if 'effusion' in abstract_text.text:
                found_effusion = True
                effusion_mention.append(xml)
                break
        except TypeError:
            continue
    if not found_effusion:
        effusion_non_mention.append(xml)
        
print('effusion mentioned in {} xmls'.format(len(effusion_mention)))   
print('effusion not mentioned in {} xmls'.format(len(effusion_non_mention)))   
    
```

### Inspect mention effusion 

```python
for xml in effusion_mention[:10]:
    root = xml['xml'].getroot()
    for pmc_id in root.iter('pmcId'):
        print('id : ', pmc_id.attrib['id'])
    
    for abstract_text in root.iter('AbstractText'):
        
        try:
            if not 'effusion' in abstract_text.text:
                continue
            print ( ' ' * 4, abstract_text.tag, abstract_text.attrib)
            print(' ' * 8, abstract_text.text + '\n')
        except TypeError:
            continue
    
    print('-' * 20)
        
    
```

## Devide positive and negative effusion

```python
negative_effusion_suspected = []
positive_effusion_suspected = []

positive_segs = []
negative_segs = []

negative_attributes = ['no ', 'No ', 'not ', 'Not ', 'without ', 'negative', 'Negative', 'clear', 'Clear']

for xml in effusion_mention:
    root = xml['xml'].getroot()
    
    positive_effusion = True
    for abstract_text in root.iter('AbstractText'):

        try:
            abs_text = abstract_text.text
            if not 'effusion' in abs_text:
                continue
            
            text_segments = abs_text.split('. ')
            
            effusion_segment = None
            for seg in text_segments:
                if 'effusion' in seg:
                    effusion_segment = seg
                    break
            
            if effusion_segment is None:
                raise ValueError('no effusion found in segments')
            
        except TypeError:
            continue
    if any([neg_atr in seg for neg_atr in negative_attributes]):
        negative_segs.append(seg)
        negative_effusion_suspected.append(xml)
    else:
        positive_segs.append(seg)
        positive_effusion_suspected.append(xml)
```

```python
print('effusion positive suspected in {} xmls'.format(len(positive_effusion_suspected)))   
print('effusion negative suspected in {} xmls'.format(len(negative_effusion_suspected)))

print('\n')
for i, seg in enumerate(positive_segs):
    print('{}. "{}" is positive'.format(i + 1, seg))
    
print('\n' + '*' * 30 + '\n')   
    
for j, seg in enumerate(negative_segs):
    print('{}. "{}" is negative'.format(j + 1, seg))


```

## document analysis

```python
import json
positive_doc_path = '/Users/Ido/Desktop/project_data/OPEN-i/data_reports_positive_suspected.json'
negative_doc_path = '/Users/Ido/Desktop/project_data/OPEN-i/data_reports_negative_suspected.json'
unmention_doc_path = '/Users/Ido/Desktop/project_data/OPEN-i/data_reports_unmentioned_plural_effusion.json'

documentation_parts = {
    'positive': {'path': positive_doc_path, 'xmls': positive_effusion_suspected},
    'negative': {'path': negative_doc_path, 'xmls': negative_effusion_suspected},
    'unmention': {'path': unmention_doc_path, 'xmls': effusion_non_mention}
                      }

parsed_abstract_texts = {}
for label, data in documentation_parts.items():
    label_abstract_texts = []
    parsed_abstract_texts[label] = label_abstract_texts
    
    for xml in data['xmls']:
        root = xml['xml'].getroot()
        sample_abstract_text = {}
        for abstract_text in root.iter('AbstractText'):
            sample_abstract_text[abstract_text.attrib['Label']] = ''
            try:
                sample_abstract_text[abstract_text.attrib['Label']] = abstract_text.text
            except TypeError:
                continue
        label_abstract_texts.append({'path': xml['path'], 'abstract_text': sample_abstract_text})
    
    with open(data['path'], 'w') as f:
        json.dump(label_abstract_texts, f)
    
    

```
